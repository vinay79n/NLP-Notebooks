### <u> Chunking <u>

### (chunking is defined as the process of NLP used to identify parts of speech and short phrases present in given sentence)
    
### (pos (part of speech) tagging is limited to identifying only singular parts of speech i.e. words but chunking is used to tagging phrases i.e small sentences)
    
### Types of Chunking

<b>There are, broadly, two types of chunking:<b>

<b>1) Chunking up<b>
    
<b>2) Chunking down<b>
    
<u><b> Chunking up <b><u>:
    
<b>Here, we dont dive deep; instead, we are happy with just an overview of the information. It just helps us get a brief idea of the given data.<b>

<u><b> Chunking down <b><u> :
    
<b>Unlike the previous type of chunking, chunking down helps us get detailed information.<b>

<b>So, if you just want an insight, consider chunking up otherwise prefer chunking down.<b>

```python
from nltk import RegexpParser
from nltk import pos_tag
```

```python
# Create a chunk grammer with on regular expression rule: 
'''According to the rule you created, your chunks:

   Start with an optional (?) determiner ('DT')
   Can have any number (*) of adjactives (JJ)
   End with a noun (<NN>)
'''
grammar = "NP: {<DT>?<JJ>*<NN>}" # DT "?" means determiner may not come then its okay but when it comes it 
                                 # must come only once
```

```python
sent = "It's a Dangerous business, Frodo, going out your door."
sent2= "It was a beautiful game by him. The amazing play!"
```

```python
chunk_parser = RegexpParser(grammar) # in chunk_parser = RegexpParser(grammar) we pass the regular expression 
                                     # grammer that we custom defined earlier 
tags = pos_tag(word_tokenize(sent2))  # we created  part-of-speech taggings of the sentence
tree = chunk_parser.parse(tags)      # created the tree using chunk_parser.parse(tags)
tree
```

## <u> Stemming <u>
    
### Stemming is the process of producing morphological variants of a root/base word. Stemming programs are commonly referred to as stemming algorithms or stemmers. A stemming algorithm reduces the words chocolates, chocolatey, choco to the root word, chocolate and retrieval, retrieved, retrieves reduce to the stem retrieve. Stemming is an important part of the pipelining process in Natural language processing. The input to the stemmer is tokenized words. How do we get these tokenized words? Well, tokenization
    
### E.g.
    
<b> Some more example of stemming for root word "like" include: <b>
    
<b>->"likes"<b>
    
<b>->"liked"<b>
    
<b>->"likely"<b>
    
<b>->"liking"<b>
    
<u><b>Errors in Stemming:<u><b>

<b>There are mainly two errors in stemming  <b>
 

<b>1) over-stemming<b> -> <b> Over-stemming occurs when two words are stemmed from the same root that are of different stems. Over-stemming can also be regarded as false-positives.<b> 
    
<b>2) under-stemming<b> -> <b>Under-stemming occurs when two words are stemmed from the same root that are not of different stems. Under-stemming can be interpreted as false-negatives.<b>

## <u>lemmatization in nlp<u>

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
```

```python
wnl = WordNetLemmatizer()
```

```python
word = 'playing'
print(wnl.lemmatize(word,'v')) # verb
print(wnl.lemmatize(word,'n')) # noun
print(wnl.lemmatize(word,'a')) # adj
print(wnl.lemmatize(word,'r')) # adverb
```

```python
words = 'players','worst','biggest','children','ate'
```

```python
for word in words:
    print(wnl.lemmatize(word,'v')) # verb
    print(wnl.lemmatize(word,'n')) # noun
    print(wnl.lemmatize(word,'a')) # adj
    print(wnl.lemmatize(word,'r')) # adverb
    print('-'*10)
```

## <U> Word Sense Disambiguation <U>
 
### In Natural Language Processing (NLP), word sense disambiguation (WSD) is the challenge of determining which sense (meaning) of a word is activated by its use in a specific context, a process that appears to be mostly unconscious in individuals.

```python
from nltk.wsd import lesk
```

```python
a1 = lesk(word_tokenize('This device is used to jam the signal'),'jam')
print(a1,a1.definition())
print('-'*50)
a2 = lesk(word_tokenize('I am stuck in traffic jam.'),'jam')
print(a2,a2.definition())
```

## <u> Lexicology <u>
    
### Lexicology is the branch of linguistics that analyzes the lexicon of a specific language. A word is the smallest meaningful unit of a language that can stand on its own, and is made up of small components called morphemes and even smaller elements known as phonemes, or distinguishing sounds. Lexicology examines every feature of a word  including formation, spelling, origin, usage, and definition.

### Lexicology also considers the relationships that exist between words. In linguistics, the lexicon of a language is composed of lexemes, which are abstract units of meaning that correspond to a set of related forms of a word. Lexicology looks at how words can be broken down as well as identifies common patterns they follow.
    
### Lexicology is associated with lexicography, which is the practice of compiling dictionaries.
