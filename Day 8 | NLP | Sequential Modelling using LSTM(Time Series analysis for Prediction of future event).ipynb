```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
alphabets
```

```python
# The enumerate() function takes a collection (e.g. a tuple) and returns it as an enumerate object. 
# The enumerate() function adds a counter as the key of the enumerate object.
list(enumerate(alphabets))
```

```python
# created a character to integer(in our case we used simple indexing) mapping database 
char_to_int = dict((c,i) for i, c in enumerate(alphabets))
char_to_int
```

```python
# created a integer(in our case we used simple indexing) to character mapping database 
int_to_char = dict((i,c) for i, c in enumerate(alphabets))
int_to_char
```

```python
seq_length = 1
```

```python
len(alphabets)
```

```python
x = []
y = []
for i in range(len(alphabets) - seq_length): # iterating form len(alphabets) - seq_length i.e. 26 - 1 = 25 
                                             # times iterations will run
        
    seq_in = alphabets[i:i + seq_length] # seq_in is list of i :  i + seq_length i.e. [0:1] -> so it'll start 
                                         # from index 0 that is with alphabet 'A'
        
    seq_out = alphabets[i + seq_length] # seq_out is list i + seq_length i.e. 0 + 1 = 1 -> so it'll start from
                                        # index 1 that is with alphabet 'B' 
        
    x.append([char_to_int[char] for char in seq_in]) # appended char_to_int values (i.e indexces) of 
                                                     # seq_in list to our x 
        
    y.append(char_to_int[seq_out]) # appended char_to_int values (i.e indexces) of seq_out list to our y
    
    print(seq_in,'->',seq_out) # printed mapping of seq_in with seq_out
```

```python
x
```

```python
y
```

```python
import numpy as np
```

```python
x = np.reshape(x, (len(x), seq_length, 1)) # reshaping x 
```

```python
x.shape
```

### Feature Scaling

```python
x = x / 25
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y
```

```python
y.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import GRU
from tensorflow.keras.layers import LSTM
```

```python
# initialized the Sequential model with LSTM layer
model = Sequential([
    Bidirectional(LSTM(32), input_shape = (x.shape[1], x.shape[2])),
    Dense(26, activation = 'softmax')
])
```

```python
#model Summary
model.summary()
```

```python
# Model Compilation
model.compile(optimizer = 'adam',
             loss = 'categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
#Fitting the model
history = model.fit(x,y,epochs = 500,batch_size = 1)
```

```python
model.evaluate(x,y,batch_size = 1)
```
