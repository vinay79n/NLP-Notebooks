### <b>Synset: a set of synonyms that share a common meaning<b>

#### <b>Each synset contains one or more lemmas, which represent a specific sense of a specific word.<b>

```python
synset = wordnet.synsets("good")
print('Word and Type : ' + synset[0].name())
print('Synonym : ' + synset[0].lemmas()[0].name())
print('The meaning of the Word : ' + synset[0].definition())
print('Example : ' + str(synset[0].examples()))
```

### <b>Each synset contains one or more lemmas(multiple synonyms and antonyms), which represent a specific sense of a specific word. and to print them we'll use forr loop<b>

```python
syn = []
ant = []
for synset in wordnet.synsets("fine"):
    for lemma in synset.lemmas():
        syn.append(lemma.name())  # add the synonyms to a list 
        if lemma.antonyms(): # if antonyms of the word are available 
                             # ,add them to a list
            ant.append(lemma.antonyms()[0].name())
print("Synonyms : ",set(syn))
print()
print("Antonyms : ",set(ant))
```

## <u>Part-Of-Speech (POS) tagging for Synonyms and Antonyms<u>

```python
word = 'match'
print("VERB : ",wordnet.synsets(word, pos = wordnet.VERB)) # part-of-speech tagging set to VERB to extract it
print()
print("ADJECTIVE : ",wordnet.synsets(word, pos = wordnet.ADJ)) # part-of-speech tagging set to ADJECTIVE to extract it
print()
print("NOUN : ",wordnet.synsets(word, pos = wordnet.NOUN)) # part-of-speech tagging set to NOUN to extract it
```

```python
# Extraction of all VERB meanings of word 'match'
syn = wordnet.synsets(word, pos = wordnet.VERB)
for x in syn:
    print(x,":",x.definition())
```

```python
# Extraction of all NOUN meanings of word 'match'
syn = wordnet.synsets(word, pos = wordnet.NOUN)
for x in syn:
    print(x,":",x.definition())
```

## <u>A Generalized code that pulls data from a web-link and returns most frequent words (below 2 cells)<u>

```python
from nltk.corpus import stopwords
swords = stopwords.words('english')

from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

from nltk import FreqDist
from nltk import pos_tag
import urllib
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
```

```python
html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Jupiter')

data = html.read()

soup = BeautifulSoup(data, 'lxml')
data = soup.get_text()

tokens = word_tokenize(data)

token1 = [x.lower() for x in tokens if x.isalpha() or x.isdigit()]
tags = pos_tag(token1)
token2 = [x[0] for x in tags if x[1].startswith('N')]

token3 = [wnl.lemmatize(x,'n') for x in token2 if x not in swords]

freq = FreqDist(token3)
freq.plot(10,cumulative=False)
```
